```python
def check_nan(df: pd.DataFrame) -> None:
    
    """
    Recibe un dataframe y enseña el % de nulos y lo grafica
    """
    
    nan_cols = df.isna().mean() * 100  # % de valores nulos
    
    nan_cols = nan_cols[nan_cols>0]
    
    display(f'N nan cols: {len(nan_cols)}')
    display(nan_cols)
    
    
    # grafico de nulos en el dataframe

    #inicializa la figura
    plt.figure(figsize=(10, 6))  # 100X60  pixeles


    sns.heatmap(df.isna(),       # datos
                yticklabels=False,  # quita las etiquetas del eje y
                cmap='viridis',      # mapa de color
                cbar=False           # sin barra lateral
               )

    plt.show();
```

Hay 1532 registros distintos lo cual es una cantidad exagera de inputs, pero mirando las distintas posibilidades creo que podemos intentar agrupar las actividades en: Surfing, Swimming, Fishing, Diving, Paddling, Standing and Boarding.

Y dejaremos las demás acepciones como "Other" o "Unknown" con el fin de poder hacer un análisis de los datos estandarizados.

```python
def replace_activity(x):
    
    activity = x.lower()
    
    if re.search(r'\bsurf', activity):
        return 'surfing'
    
    elif re.search(r'\bswim', activity):
        return 'swimming'
    
    elif re.search(r'\bfish', activity):
        return 'fishing'
    
    elif re.search(r'\bdiv', activity):
        return 'diving'
    
    elif re.search(r'\bpaddl', activity):
        return 'paddling'
    
    elif re.search(r'\bstand', activity):
        return 'standing'
    
    elif re.search(r'\bboard', activity):
        return 'boarding'
    
    elif activity == 'unknown':
        return 'unknown'
    
    else:
        return 'other'
```

```python
df.Activity = df.Activity.apply(replace_activity)
```

```python
activity = df.Activity.value_counts()
activity
```

```python
activity_counts = activity[activity.index != 'other']
```

Mi filtro ha funcionado, pero devido a los distintos inputs que se encontraban, la mayoría se encuentra ahora en 'other', pero como no es un valor que pueda standarizar o que sea útil, porque puede ser cualquier cosa, lo mejor es centrarnos en los demás valores que sí son interesante.

```python
activity_counts.plot(kind='bar');
```

Como podemos ver existe una mayor cantidad de ataques en aquellas personas que surfean y que nadan. 
Como decíamos al principio, mi hipótesis sostiene que los tiburones no atacan a las personas de manera indiscriminada, si no que muchas veces los confunden con otros animales.

```python
def timeTable(x):
    
    timelst = ['morning','afternoon','evening','night']
    

    try:
        if x in timelst:
            return x

        elif '-' in x:

            y = x.split('-')
            z = []
            for i in y:
                z.append(float(i))

            z2 = np.mean(z)

            if z2 >= 5.00 and z2 <= 11.59:

                return 'morning'

            elif z2 >= 12.00 and z2 <= 17.59:
                return 'afternoon'

            elif z2 >= 18.00 and z2 <= 20.59:
                return 'evening'

            elif z2 >= 21.00 and z2 <= 4.59:
                return 'night'

        elif '.' in x:
            x = float(x)
            if x >= 5.00 and x <= 11.59:

                return 'morning'

            elif x >= 12.00 and x <= 17.59:
                return 'afternoon'

            elif x >= 18.00 and x <= 20.59:
                return 'evening'

            elif x >= 21.00 and x <= 4.59:
                return 'night'
    except:
        return x
```

```python
df.Time = df.Time.apply(timeTable)
```

```python
def sharks(x):
    
    try:
        shark = x.lower()

        if re.search(r'\bwhite', shark):
            return 'White Shark'

        elif re.search(r'\btiger', shark):
            return 'Tiger Shark'

        elif re.search(r'\blemon', shark):
            return 'Lemon Shark'

        elif re.search(r'\bbull', shark):
            return 'Bull Shark'

        elif re.search(r'\bgrey reef', shark):
            return 'Grey Reef Shark'

        elif re.search(r'\bNurse', shark):
            return 'Nurse Shark'

        elif re.search(r'\bwobbegong', shark):
            return 'Wobbegong Shark'

        elif re.search(r'\bblacktip', shark):
            return 'Blacktip Shark'

        elif re.search(r'\bspinner', shark):
            return 'Spinner Shark'

        elif re.search(r'\bblue', shark):
            return 'Blue Shark'

        elif re.search(r'\bmako', shark):
            return 'Mako Shark'

        elif re.search(r'\bbronze', shark):
            return 'Bronze Shark'

        elif re.search(r'\bhammerhead', shark):
            return 'Hammerhead Shark'

        elif re.search(r'\bgoblin', shark):
            return 'Goblin Shark'

        elif re.search(r'\bsevengill', shark):
            return 'Sevengill Shark'

        elif shark == 'unknown':
            return 'unknown'

        else:
            return 'other'
    except:
        pass
```

```python
df['Species'] = df['Species'].apply(sharks)
```
